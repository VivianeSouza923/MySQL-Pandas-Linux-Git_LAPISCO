```python
# processo de limpeza da base (remover parenteses, colunas desnecessárias etc.)

df_titanic = pd.read_csv('titanic.csv', usecols=['PassengerId', 'Name', 'Sex', 'Age', 'Survived'])

df_titanic.head()
```

```python
def remove_parenteses(item):
  if '(' in item:
    return item.replace('(','').replace(')','')
  else:
    return item

df_titanic.Name.head(10).apply(remove_parenteses)
```

```python
df_titanic
```

```python
# rstrip() remove espaços à direita

df_titanic.Name = df_titanic.Name.str.rstrip()
df_titanic
```

```python
# lstrip() remove espaços à esquerda

df_titanic.Name = df_titanic.Name.str.lstrip()
df_titanic
```

```python
# upper() transforma o texto em maiúsculo

df_titanic.Name = df_titanic.Name.str.upper()
df_titanic
```
