0 a 18 anos → Criança

18 a 65 anos → Adulto

maior de 65 → Idoso


```python
Titanic_for_age = df_titanic.values.tolist()
# Titanic_for_age

for i, a in enumerate(Titanic_for_age):
    if a[4] <= 18:
        Titanic_for_age[i][4] = 'Criança'
    elif a[4] > 18 and a[4] <= 65:
        Titanic_for_age[i][4] = 'Adulto'
    elif a[4] > 65:
        Titanic_for_age[i][4] = 'Idoso'
    

Titanic_final = pd.DataFrame(Titanic_for_age, columns=['PassengerId','Survived','Name', 'Sex', 'Age'])
Titanic_final
        
```
