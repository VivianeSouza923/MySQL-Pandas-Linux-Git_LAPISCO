# Consultando dados com o query

* semelhante as queries do SQL
* operadores lógicos AND(&) e OR( | )
* expressões lógicas (==, != , >, <, >= e <=)

```python
# pegar todas as idades maiores que 20 e retornar os 6 primeiros valores
# df.query(condição que desejo).head(x)
# explicando a linha acima: eu quero todos os dados que tem na coluna 'age' números maiores que 20 e quero
# que seja exibido os 6 primeiros casos que atendem a essa condição
df.query('Age > 20').head(6)
```

```python
# pegar todas as idades maiores que 20 de pessoas do sexo masculino e retornar os 10 primeiros valores. "E - AND" - Operador lógico
# df.query('condição1 & condição2').head(x)
# and ---> &
df.query('Age > 20 & Sex=="male"').head(10)
```

```python
# pegar todas as idades maiores que 20 ou pessoas do sexo masculino e retornar os 10 primeiros valores. "OU - OR" - Operador lógico
# df.query('condição1 | condição2').head(x)
# or --> |
df.query('Age > 20 | Sex=="male"').head(10)
```

## Métodos de preenchimento de valores faltantes.

* df.fillna(0)
* df.fillna(df.Age.mean())
* df.fillna(method='ffill') # forward fill
* df.fillna(method='bfill') # backward fill
* values = {'A': 0, 'B': 1, 'C': 2, 'D': 3} -> df.fillna(value=values)


### Preencher os valores faltantes da coluna ‘Age’ e da coluna ‘Embarked’ com a moda. Para coluna ‘Cabin’, substituir todos os valores faltantes pela sigla SC (Sem Cabine).

```python
# aqui eu vou criar um dicionário (lembre-se que ele vem dentro de chaves)
# para a idadade eu quero a moda, para cabin eu vou adicionar SC e para Embarked eu quero a moda também
values = {'Age': df.Age.mode()[0], 'Cabin': 'SC', 'Embarked': df.Embarked.mode()[0]}
print(values)
```

```python
# Atribuido os novos valores
# utiliza-se essa função fillna que tem como parametro value que irá receber o dicionário que nós criamos na célula
# acima. Colocamos o inplace=True para alterar o df
df.fillna(value=values, inplace=True)
```

```python
# vendo como ficou quando atribuimos os novos valores
# vemos a soma dos elementos nulos em cada coluna
df.isnull().sum()
```

```python
# aqui eu vou criar a variável teste qye iraá o receber um dataframe com os elementos que eu quero
# são 3 linhas e 3 colunas (3x3)
# só usar a pd.DataFrame([[],[],[]])
teste = pd.DataFrame([[1, 2, 3], [4, None, None], [None, None, 9]])
teste
```

```python
# forward fill
# onde ficou nulo, será preenchido com o que vem de cima. Como isso acontece?
# por exemplo, o lugar 20 está nulo, portanto ele será preenchido com o valor que vem de cima, ou seja, o lugar 10
# só usar a função fillna(). Como ela funciona?
# dataframe.fillna(method='o que eu quero ---> no nosso contexto, será o ffill')
teste.fillna(method='ffill')
```

```python
# backward fill
# aqui teremos o inverso da célula acima: quero preencher onde tá nulo com o que vem abaixo dele.
# por exemplo, o lugar 12 está vazio, portanto ele será preenchido com o que está no lugar 22
# só usar a função fillna(). como ela funciona?
# dataframe.fillna(method='bfill')
teste.fillna(method='bfill')
```

### Cleaning textos

```python
# Remove todos esses parênteses na coluna 'Name'
# explicando: função normal que a gente cria em Python. Cria a função com o parametro de entrada (no nosso caso, é
# o "item") e para qualquer '(' dentro desse item, eu vou retorná-lo após uma aplicação de replace ('antigo', 'novo')
# depois disso é só applicar essa função em Name, através do apply()
def remove_parenteses(item):
  if '(' in item:
    return item.replace('(','').replace(')','')
  else:
    return item

df.Name.head(10).apply(remove_parenteses)
```

```python
# continua aquiiiii 29/07/2021
# PARA ISSO OBSERVE A COLUNA Dtype, que é onde diz o tipo de memória daquele dado!

# tipo objeto consome muita memória.  OBserve Memory usage agora.
df = pd.read_csv("titanic.csv")
df.info()
```

#### A coluna ‘Sex’ foi classificada como object, mas ela pode ser convertida para ‘category’, pois o sexo na verdade é uma característica. Com essas conversões conseguimos diminuir o consumo de memória.


```python
# o tipo de dado category consome menos memória do que o tipo objeto.
# Convertendo os tipos de dados:

#observe agora que conseguimos diminuir a memória utilizada

# vou armazenar a mudança de tipo de dados só na coluna do próprio dataframe (se eu quero mudar o Pclass, salvo como
# df.Pclass, entendeu?)

# agora só usar o astype('aqui dentro vem o nome da nova categoria')
# df.Vivi = df.Vivi.astype('Viviane') --> exemplinho

df.Sex = df.Sex.astype('category')
df.Embarked = df.Embarked.astype('category')
df.Survived = df.Survived.astype('category')
df.Pclass = df.Pclass.astype('category')
df.PassengerId = df.PassengerId.astype('int32')
df.Parch = df.Parch.astype('int32')
df.SibSp = df.SibSp.astype('int32')

df.info()
```
